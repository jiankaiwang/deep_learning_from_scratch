## ELement-wise Calculation

```python
W = np.array([[1,2,3], [4,5,6]])
X = np.array([[0,1,2], [3,4,5]])
```

```python
W+X
```

```python
W*X
```

## Broadcast

```python
A = np.array([[1,2],[3,4]])
print(A * 10)
```

```python
b = np.array([10,20])
print(A * b)
```

## Inner Product

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(np.dot(a,b))
```

## Matrix Multiply

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(np.dot(A,B))
```
